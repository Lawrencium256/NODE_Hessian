```python
import numpy as np


npoints = 40           #Number of points in the data set.
ninner = 20           #Number of points in the inner circle
nouter = 20           #and the outer ring
dim = 3
test_train = 'test' #choose train or test to choose the name
r = np.array([[0, 0.3, 0.7, 1], [0, 0.5, 0.85, 1], [0, 0.3, 0.7, 1], [0, 0.3, 0.7, 1]])
r_sq = r[dim-1]**2

name_in = ('/content/drive/MyDrive/colab_notebooks/NODE_Hessian/nested_spheres/cross_entropy/experiment_0/data/' 
           + str(dim) +'din_'+str(npoints)+'_'+test_train+'.npy')
name_out = ('/content/drive/MyDrive/colab_notebooks/NODE_Hessian/nested_spheres/cross_entropy/experiment_0/data/' 
            + str(dim)+'dout_'+str(npoints)+'_'+test_train+'.npy')

in_data = np.empty((npoints, dim))
out_data = np.empty((npoints, 1))
```

```python
i = 0
while i < ninner:
    point = np.random.rand(dim)*2 - 1            #Random array of numbers with the correct shape. 
                                                 #Does so from a uniform distribution over [0, 1). The elements of point can take 
                                                 #any value between 1 and -1. This essentially produces a random point that 
                                                 #lies within a circle/n-sphere of radius 1, centred on the origin.
    dot = np.dot(point, point)                   #Inner product of the point with itself. This gives the radius.
    if (r_sq[0] <= dot) and (dot <= r_sq[1]):    #i.e. if the point is within the inner region.
        in_data[i] = point
        out_data[i] = [0]                        #The point has label 0.
        i += 1
    else:                                        #Else, just ignore this and try again.
        pass
  
i = 0    
while i < nouter:                                
    point = np.random.rand(dim)*2 - 1
    dot = np.dot(point, point)
    if (r_sq[2] <= dot) and (dot <= r_sq[3]):   #i.e. if the point is within the outer region.
        in_data[i+ninner] = point
        out_data[i+ninner] = [1]
        i += 1                                  #The point has label 1.
    else:
        pass
```

```python
np.save(name_in, in_data)
np.save(name_out, out_data)
```
